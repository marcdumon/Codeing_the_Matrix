# UN Voting Analysis

```python
import pandas as pd
```

```python
f = open('UN_voting_data.txt', 'r')
voting_data = []
for row in list(f):
    voting_data.append(row.split(' '))
voting_data = pd.DataFrame(voting_data)
# Remove \n from last column
voting_data.iloc[:,-1] =  voting_data.iloc[:,-1].str.split(expand=True)
# convert str to int
voting_data.iloc[:,1:] = voting_data.iloc[:,1:].astype(int)
```

```python
voting_data
```

```python
get_votes = lambda country: voting_data[voting_data.iloc[:, 0]==country].iloc[:,1:].values[0]
calc_similarity = lambda country1, country2: sum(get_votes(country1) * get_votes(country2)) 
```

```python
for row in voting_data
```

```python
calc_similarity('Belgium', 'Belgium'), calc_similarity('Belgium', 'Netherlands'), calc_similarity('Belgium', 'France'), calc_similarity('Belgium', 'Germany')
```

```python
cross_similarity = pd.DataFrame(columns=voting_data[0])
for country1 in voting_data[0]:
    for country2 in voting_data[0]:
        cross_similarity.loc[country1,country2] = calc_similarity(country1, country2)
```
